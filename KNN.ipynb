```python
# import library
import numpy as np              # numpy digunakan untuk operasi numerik
from collections import Counter # Counter digunakan untuk menghitung frekuensi elemen dalam sebuah list.

#mendefinisikan kelas "KNN"
class KNN:
    def __init__(self, k=3):    # Metode __init__ ini menginisialisasi objek KNN dengan parameter opsional k=3
        self.k = k              # X_train dan y_train diatur ke None untuk menyimpan data training setelah model dilatih.
        self.X_train = None
        self.y_train = None
#training data
    def fit(self, X_train, y_train):    # Metode ini digunakan untuk menyimpan data training ke dalam objek KNN.
        self.X_train = X_train          # menerima data latihan X_train dan label y_train, lalu menyimpannya dalam atribut X_train dan y_train
        self.y_train = y_train          # untuk digunakan nanti.
#melakukan prediksi
    def predict(self, X):                             # predict ini adalah Method untuk melakukan prediksi pada data uji.
        predictions = [self._predict(x) for x in X]   # Mengambil array X sebagai parameter, yang berisi fitur-fitur data uji.
        return predictions

    def _predict(self, x): # _predict method ini melakukan langkah-langkah utama algoritma KNN untuk satu titik data uji (x):
        # Kode dibawah ini digunakan untuk Menghitung jarak antara titik data uji x dengan setiap titik data pada data training menggunakan Euclidean distance.
        jarak = [np.linalg.norm(x - x_train) for x_train in self.X_train]

        # menghasilkan array indeks yang menunjukkan urutan k tetangga terdekat berdasarkan jarak dari yang terdekat hingga yang terjauh.
        k_neighbors_indices = np.argsort(jarak)[:self.k]

        # Mengambil label dari 'k' tetangga terdekat
        k_neighbor_labels = [self.y_train[i] for i in k_neighbors_indices]

        # Menentukan kelas mayoritas di antara tetangga terdekat
        most_common = Counter(k_neighbor_labels).most_common(1)
        return most_common[0][0]
```

```python
#contoh penggunaan

if __name__ == "__main__":
    # Data training
    X_train = np.array([[1.2, 2.3],[2.5, 4.6],[4, 1],[5.6, 1.2],[6, 3.5]])
    y_train = np.array(['A','A','B','B','B'])

    # Data test
    X_test = np.array([[3, 2]])

    # Pada bagian akhir kode, dilakukan contoh penggunaan dengan membuat objek "klasifikasi_knn" dari kelas "KNN"
    klasifikasi_knn = KNN(k=3)
    klasifikasi_knn.fit(X_train, y_train)

    #Menerapkan model KNN yang telah dilatih (klasifikasi_knn) untuk melakukan prediksi pada data testing (X_test).
    #Kemudian hasil prediksi disimpan dalam variabel predictions.
    predictions = klasifikasi_knn.predict(X_test)
```

```python
    # mencetak hasil prediksi
    print("Kelas:", predictions)
```
